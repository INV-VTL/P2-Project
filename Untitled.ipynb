```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
```

```python
f1 = 'price~sqft_living'
model_1 = smf.ols(formula=f1, data=house_df).fit()
model_1.summary()
```

```python
f2 = 'price~sqft_above'
model_2 = smf.ols(formula=f2, data=house_df).fit()
model_2.summary()
```

```python
f3 = 'price~sqft_living15'
model_3 = smf.ols(formula=f3, data=house_df).fit()
model_3.summary()
```

```python
x_new = pd.DataFrame({'sqft_living': [house_df.sqft_living.min(), house_df.sqft_living.max()]})
preds = model_1.predict(x_new)
```

```python
house_df.plot(kind = 'scatter', x = 'sqft_living', y = 'price')
plt.plot(x_new, preds, c = 'red', linewidth = 2)
plt.show()
```

```python
fig1 = plt.figure(figsize = (15, 8))
fig1 = sm.graphics.plot_regress_exog(model_1, "sqft_living", fig = fig1)
plt.show()
```

```python
residuals = model_1.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```
